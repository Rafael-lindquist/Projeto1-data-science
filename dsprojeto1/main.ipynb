```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Relatorio Estagios até março 2024 - Planilha1.csv")
df = df.rename(columns=df.iloc[0]).drop(df.index[0])
df.drop(columns="#", axis=1, inplace=True)
df
```

```python
# 1) Gráfico do número de estágios por porcentagem concluida do curso

from typing import Counter
dc = df['Percentual de Disciplinas Obrigatórias Cursadas']
dc = [float(i) for i in dc]
dc = [int(i) for i in dc]
dc = Counter(dc)

# Visualização
plt.figure(figsize=(15, 5))
plt.bar([i for i in dc.keys()], [i for i in dc.values()])
plt.title("Número de estágios por porcentagem concluida do curso")
plt.ylabel("Número de estágios")
plt.xlabel("Porcentagem concluida do curso")
plt.show()
```

```python
# 2) Quais cursos tem mais estágios

cursos = df["Curso"]
cursos = Counter(cursos)
cursos = [(i, j) for i, j in cursos.items()]
cursos = sorted(cursos, key=lambda x: x[1], reverse=True)
x, y = [i[0] for i in cursos[:10]], [i[1] for i in cursos[:10]]

# Visualização
plt.figure(figsize=(15, 5))
plt.bar(x, y)
plt.title("Cursos com mais estágios")
plt.xlabel("Cursos")
plt.xticks(fontsize=8, rotation=90)
plt.ylabel("Número de estágios")
plt.title("Os 10 cursos com mais estágios")
plt.show()

# Análise
quatromaiores = [i for i in df['Curso']
                  if i == 'TECNOLOGIA EM ANÁLISE E DESENVOLVIMENTO DE SISTEMAS' 
                  or i == 'BACHARELADO EM ENGENHARIA MECÂNICA'
                  or i == 'BACHARELADO EM ENGENHARIA DE CONTROLE E AUTOMAÇÃO'
                  or i == 'LICENCIATURA EM PEDAGOGIA']
quatromaiores = len(quatromaiores)
percentualquatromaiores = quatromaiores / len(df['Curso'])
print(percentualquatromaiores)
```

```python
# 3) Cidades com mais estágios
cidades = df["Concedente Cidade"].dropna()
cidades = Counter(cidades)
cidades = [(i, j) for i, j in cidades.items()]
cidades = sorted(cidades, key=lambda x: x[1], reverse=True)
x, y = [i[0] for i in cidades[:10]], [i[1] for i in cidades[:10]]

# Visualização
plt.figure(figsize=(15,5))
plt.bar(x, y)
plt.xlabel("Cidades")
plt.xticks(fontsize= 8, rotation=30)
plt.ylabel("Número de estágios")
plt.title("As 10 cidades com mais estágios")
plt.show()

saopaulo = [i for i in df["Concedente Cidade"] if i == "SÃO PAULO / SP"]
porcentagemsaopaulo = len(saopaulo) / len(df['Concedente Cidade'])
print(porcentagemsaopaulo)
```
